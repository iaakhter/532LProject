```python
import os

def create_list(root_dir, listfile):
    w = open(listfile, "w")
    genres = sorted(os.listdir(root_dir))
    for i, genre in enumerate(genres):
        genre_dir = os.path.join(root_dir, genre)
        songs = os.listdir(genre_dir)
        for song in songs:
            song_full_dir = os.path.join(genre_dir, song)
            w.write("%s %d\n" % (song_full_dir, i))
```

```python
train_dir = '/ubc/cs/research/tracking-raid/candice/project/dataset/mag_512/train'
train_list = '/ubc/cs/research/tracking-raid/candice/project/dataset/mag_512/lists/train_list.txt'
create_list(train_dir, train_list)
```

```python
test_dir = '/ubc/cs/research/tracking-raid/candice/project/dataset/mag_512/val'
test_list = '/ubc/cs/research/tracking-raid/candice/project/dataset/mag_512/lists/val_list.txt'
create_list(test_dir, test_list)
```

```python
from random import shuffle

train_list = "/ubc/cs/research/tracking-raid/candice/project/dataset/mag_512/lists/train_list.txt"
test_list = "/ubc/cs/research/tracking-raid/candice/project/dataset/mag_512/lists/val_list.txt"
new_test_list = "/ubc/cs/research/tracking-raid/candice/project/dataset/mag_512/lists/new_val_list.txt"
new_train_list = "/ubc/cs/research/tracking-raid/candice/project/dataset/mag_512/lists/new_train_list.txt"
w_train = open(new_train_list, "w")
w_test = open(new_test_list, "w")
train_lines = [line.rstrip('\n') for line in open(train_list)]
test_lines = [line.rstrip('\n') for line in open(test_list)]
shuffle(test_lines)

for line in train_lines:
    print "here"
    w_train.write("%s\n" % line)
    
for i, line in enumerate(test_lines):
    if i < len(test_lines) // 3:
        w_test.write("%s\n" % line)
    else:
        w_train.write("%s\n" % line)
```
